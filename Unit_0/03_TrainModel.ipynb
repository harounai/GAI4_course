```python
import timm
import torch
import torch.nn as nn

# 'efficientnet_b0' is the smallest standard variant
# pretrained=True downloads weights optimized on ImageNet
model = timm.create_model('efficientnet_b0', pretrained=False, num_classes=2)

# Verify the output features of the last layer
print(f"Model: {model.default_cfg['architecture']}")
print(f"Classifier head: {model.get_classifier()}")
```

```python
import os
from pathlib import Path

# Check current working directory
print("Current working directory:", os.getcwd())

# List contents of the training class folders
train_class0 = Path("PrepData/Training/Class_0")
train_class1 = Path("PrepData/Training/Class_1")
print(f"Class_0 exists? {train_class0.exists()}, files: {list(train_class0.glob('*'))[:5]}")
print(f"Class_1 exists? {train_class1.exists()}, files: {list(train_class1.glob('*'))[:5]}")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
```

```python
print("Current working directory:", os.getcwd())

for split in ["Training", "Validation"]:
    for class_name in ["Class_0", "Class_1"]:
        folder = Path(f"PrepData/{split}/{class_name}")
        if folder.exists():
            files = list(folder.glob("*"))
            print(f"{folder}: {len(files)} items")
            # Show first 5 file names and extensions
            for f in files[:5]:
                print(f"  {f.name} (ext: {f.suffix})")
        else:
            print(f"{folder} does not exist")
```
